```python
def lorenzatt(X, rho=28.0, sigma=10.0, beta=3.0/8.0):
    '''
    Lorenz Attractor equations (to be solved with scipy.integrate.odeint)
    
    x' = sigma*(y-x)
    y' = x*(rho - z) - y
    z' = x*y - beta*z
    '''
    
    dx = np.zeros((3), 'd')
    dx[0] = sigma*(X[1] - X[0]);
    dx[1] = X[0]*(rho - X[2]) - X[1]
    dx[2] = X[0]*X[1] - beta*X[2]
    return dx


class L40(object):
    '''Lorenz 40 model of zonal atmospheric flow'''
    
    def __init__(self, members=1, n=40, dt=0.05, F=8):  
        self.n = n
        self.dt = dt
        self.dtx = dt
        self.x = np.random.normal(0., 0.1, size=(members, n))
        self.members = members
        self.F = F
    
    def dxdt(self):
        dxdt = np.zeros((self.members, self.n),'f8')
        for n in range(2,self.n-1):
            dxdt[:,n] = -self.x[:,n-2]*self.x[:,n-1] +  \
                        self.x[:,n-1]*self.x[:,n+1] - self.x[:,n] + self.F
        dxdt[:,0] = -self.x[:,self.n-2]*self.x[:,self.n-1] +  \
                self.x[:,self.n-1]*self.x[:,1] - self.x[:,0] + self.F
        dxdt[:,1] = -self.x[:,self.n-1]*self.x[:,0] + \
                self.x[:,0]*self.x[:,2] - self.x[:,1] + self.F
        dxdt[:,self.n-1] = -self.x[:,self.n-3]*self.x[:,self.n-2] + \
                            self.x[:,self.n-2]*self.x[:,0] - \
                            self.x[:,self.n-1] + self.F
        return dxdt
    
    def rk4step(self):
        h = self.dt; hh = 0.5*h; h6 = h/6.
        x = self.x
        dxdt1 = self.dxdt()
        self.x = x + hh*dxdt1
        dxdt2 = self.dxdt()
        self.x = x + hh*dxdt2
        dxdt = self.dxdt()
        self.x = x + h*dxdt
        dxdt2 = 2.0*(dxdt2 + dxdt)
        dxdt = self.dxdt()
        self.x = x + h6*(dxdt1 + dxdt + dxdt2)
    
  
    def assimilate(self,yo,H,obs_std=1,members=50): 
        # this is the function of square root filter 
        # yo:obsearvation
        # obs_std: observation standard diviation 
        for i in range(len(yo)):
            yoi=yo[i]
            hi=H[i]

            xb_bar=self.x.mean(axis=0)
            xb_prime=self.x-xb_bar
            
            Pb=np.sum((xb_prime[:,:,np.newaxis])*(xb_prime\
                                            [:,np.newaxis,:]),axis=0)/(members-1)
            R = obs_std**2*np.eye(len(yo))
            
            K=np.dot(np.dot(Pb,H.T),np.linalg.inv(np.dot(np.dot(H,Pb),H.T)+R))
            
            alpha=(1+np.sqrt((R[0,0]/(np.dot(np.dot(hi,Pb),hi.T)+R[0,0]))))**(-1)
            
        
            xbar_a=xb_bar+np.dot(K,(yo-np.dot(H,xb_bar)))
            
            xprime_a=np.zeros_like(xb_prime)
            xa=np.zeros_like(xb_prime)
            Kt=alpha*np.dot(np.dot(Pb,hi.T),1./(np.dot(np.dot(hi,Pb),hi.T)+R[0,0]))
            for j in range(members):
                xprime_a[j]=xb_prime[j]+np.dot(Kt,(-np.dot(hi,xb_prime[j])))
                xa[j]=xbar_a+xprime_a[j]
           
            self.x=xa
    
    def get_obs(self,H,obs_std,**kw):
        # this is the function to get observations from true value
        # obs: 
        yo=np.dot(H,self.x[0])+np.random.randn(H.shape[0])*obs_std**2
        
            
        return yo
    def assimilate2(self,yo,H,obs_std=1,members=50):
        # this is the function of perturbation 
        # yo:obsearvation
        # obs_std: observation standard diviation 
        xb_bar=self.x.mean(axis=0)
        xb_prime=self.x-xb_bar
    
        Pb=np.sum((xb_prime[:,:,np.newaxis])*(xb_prime\
                                            [:,np.newaxis,:]),axis=0)/(members-1)

        R=np.eye(H.shape[0])*obs_std**2
        K=np.dot(np.dot(Pb,H.T),np.linalg.inv(np.dot(np.dot(H,Pb),H.T)+R))
        #Pa=np.dot((1-K),Pb)
        
        xbar_a=xb_bar+np.dot(K,(yo-np.dot(H,xb_bar)))
        xprime_a=np.zeros_like(xb_prime)
        xa=np.zeros_like(xb_prime)
        yprime=np.random.randn(len(yo))*obs_std**2
        
        for i in range(members):
            xprime_a[i]=xb_prime[i]+np.dot(K,yprime-np.dot(H,xb_prime[i]))
            xa[i]=xbar_a+xprime_a[i]
        self.x=xa
    
    def save(self,s):
        if len(s)==0:
            return np.nanmean(self.x,axis=0)[np.newaxis,:]
        else:
            return np.append(s,np.nanmean(self.x,axis=0)[np.newaxis,:],axis=0)
```

```python
def run_assis(m,ens,cycle,obs,k_type,h_type):
    # this is the function that run 1000 steps and assimilate data
    # m: true value 
    # my_m: get true value and run forward without changing value of m
    # ens: ensemble 
    # cycle: assilate date every cycle steps
    # obs: number of observation
    # k_type: kt: square root filter; k: perturbation filter
    # h_type: i: simple H ;f: fourier H
    my_m=L40()
    my_ens=L40(members=50)
    my_m.x=m.x
    my_ens.x=ens.x
    run=1000
    error1=[]
    os=1
    if h_type=='f':
        H=np.zeros([m.n,m.n])
        for i in range(m.n):
            for j in range(m.n):
                H[i,j]=np.cos(j*np.pi/39.*(i))
    if h_type=='i':
        H=np.eye(m.n,m.n)
    if obs==20:
        H=H[::2,:]
    for i in range(run):
        my_m.rk4step()
        my_ens.rk4step()

        if np.mod(i,cycle)==0:
            yo=my_m.get_obs(H,os)
            if k_type=='k':
                my_ens.assimilate2(yo,H,os)
            if k_type=='kt':
                my_ens.assimilate(yo,H,os)
        if len(error1)==0:
            error1=np.sqrt((my_m.x[0]-np.nanmean(my_ens.x,axis=0))**2)[np.newaxis,:]
        else:
            error1=np.append(error1,(np.sqrt((my_m.x[0]-np.nanmean(my_ens.x,axis=0))**2)\
                               [np.newaxis,:]),axis=0)
    return np.nanmean(error1,axis=1)
```

```python
members=50
ens=L40(members=50)
m=L40()
ens.x=m.x[0]+0.01*np.random.randn(members,m.n)
for i in range(100):
    m.rk4step()
    ens.rk4step()
```

```python
error1=run_assis(m,ens,cycle=10,obs=40,k_type='kt',h_type='i')
error2=run_assis(m,ens,cycle=3,obs=40,k_type='kt',h_type='i')
error3=run_assis(m,ens,cycle=1,obs=40,k_type='kt',h_type='i')
```

```python
error7=run_assis(m,ens,cycle=10,obs=40,k_type='k',h_type='i')
error8=run_assis(m,ens,cycle=3,obs=40,k_type='k',h_type='i')
error9=run_assis(m,ens,cycle=1,obs=40,k_type='k',h_type='i')
```
